```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.cluster.hierarchy import dendrogram, linkage 
import scipy.cluster.hierarchy as shc 
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv("C:\\Users\\Sandeep\\OneDrive\\Desktop\\Classes and Files\\shopping_data.csv")
```

```python
# Taking 2 fratures Age and Anual Income
```

```python
data = df.iloc[:,2:4]
```

```python
data.head()
```

```python
# Staring the Clustering
```

```python
plt.figure(figsize=(12,8))
shc.dendrogram(shc.linkage(data, method='ward'));
```

```python
# upto here we got a basic idea about how many cluster should be there
# In this case 4 clusters is a good option
```

```python
cluster = AgglomerativeClustering(n_clusters= 4, affinity = 'euclidean', linkage = 'ward')
```

```python
cluster.fit_predict(data)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(data.iloc[:,0], data.iloc[:,1], c = cluster.labels_, cmap = 'rainbow')
plt.xlabel('Age')
plt.ylabel('Anual income')
```
