```python
X = Db[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = Db[['Outcome']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size = 0.7, random_state = 123)
```

```python
print("Shapes")
print("X_train",X_train.shape)
print("y_train",y_train.shape)
print("X_test",X_test.shape)
print("y_test",y_test.shape)
```

# Boosting 

## 1) XgBoost

```python
from xgboost import XGBClassifier
```

```python
xgb =XGBClassifier()                             # for Parameters we got same stuff as RF i.e. N_estimator, Depth , Criteria
xgb.fit(X_train, y_train)                        # We also have learning rate
                                                 # By default the depth = 6
                                                 #                N_estimator = 100
                                                 #                learning rate = 0.3
```

```python
pred_train = xgb.predict(X_train)

print("Confusion matrix")
print(confusion_matrix(pred_train, y_train))
print(" ")
print("Classification Report")
print(classification_report(pred_train, y_train))
```

```python
xgbt =XGBClassifier() 
xgbt.fit(X_test, y_test)
```

```python
pred_test = xgb.predict(X_test)

print("Confusion matrix")
print(confusion_matrix(y_test, pred_test))
print(" ")
print("Classification Report")
print(classification_report(y_test, pred_test))
```

```python
xgb1 =XGBClassifier(max_depth = 4)                # Since the model is overfitted lets reduce the max depth to make the model simpler
xgb1.fit(X_train, y_train) 
```

```python
pred_train1 = xgb1.predict(X_train)

print("Confusion matrix")
print(confusion_matrix(y_train, pred_train1))
print(" ")
print("Classification Report")
print(classification_report(y_train, pred_train1))
```

```python
xgb_t =XGBClassifier(max_depth = 4)
xgb_t.fit(X_test, y_test)
```

```python
pred_test_1 = xgb_t.predict(X_test)

print("Confusion matrix")
print(confusion_matrix(y_test, pred_test_1))
print(" ")
print("Classification Report")
print(classification_report(y_test, pred_test_1))
```

```python
xgb2 =XGBClassifier(max_depth = 5, criterion = 'entropy', n_estimators = 200) 
xgb2.fit(X_train, y_train)
```

```python
pred_train2 = xgb2.predict(X_train)

print("Confusion matrix")
print(confusion_matrix(y_train, pred_train2))
print(" ")
print("Classification Report")
print(classification_report(y_train, pred_train2))
```

```python
xgb_2 =XGBClassifier(max_depth = 5, criterion = 'entropy', n_estimators = 200) 
xgb_2.fit(X_test, y_test)
```

```python
pred_test2 = xgb2.predict(X_test)

print("Confusion matrix")
print(confusion_matrix(y_test, pred_test2))
print(" ")
print("Classification Report")
print(classification_report(y_test, pred_test2))
```
